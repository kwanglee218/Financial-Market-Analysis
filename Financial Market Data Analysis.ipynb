```python
df_sp500 = pd.concat([df_2002,df_2012,df_2007,df_2017,df_2008,df_2018,df_2009,df_2019,df_2006,
                     df_2016,df_2003,df_2013,df_2014,df_2004,df_2021,df_2011,df_2001,df_2010,df_2000,df_2015,df_2022,df_2020,df_2005])
```

```python
df_sp500['Year'] = df_sp500['Date'].str.split('/', expand=True)[2]
df_sp500['Month'] = df_sp500['Date'].str.split('/', expand=True)[0]
df_sp500['Year_Month'] = df_sp500['Year'] + '_' + df_sp500['Month']
df_sp500['Category'] = 'S&P500'
df_sp500['Date'] = pd.to_datetime(df_sp500['Date'], format='%M/%d/%Y')
df_sp500['Date'] = df_sp500['Date'].dt.strftime('%Y-%M-%d')
df_sp500['Close'] = df_sp500['Close'].str.replace(',','').astype(float)
df_sp500 = df_sp500[['Date','Year','Month','Year_Month','Close','Category']]
df_sp500 = df_sp500.sort_values(['Year','Date'])
df_sp500 = df_sp500.set_index('Date')
df_sp500_2019 = df_sp500[df_sp500['Year'] >= '2019']
df_sp500_2008 = df_sp500[df_sp500['Year'] >= '2008']
df_sp500_2008.head()
```

# USD vs. KRW Currency Rate

```python
df_currency = pd.read_csv('USD_KRW.csv')

df_currency['Year'] = df_currency['Date'].str.split('/', expand=True)[2]
df_currency['Year'].replace({'22':'2022','21':'2021', '20':'2020', '19':'2019', '18':'2018', '17':'2017',
                             '16':'2016', '15':'2015', '14':'2014', '13':'2013', '12':'2012',
                             '11':'2011', '10':'2010', '09':'2009', '08':'2008', '07':'2007'}, inplace=True)
df_currency['Month'] = df_currency['Date'].str.split('/', expand=True)[0]
df_currency['Year_Month'] = df_currency['Year'] + '_' + df_currency['Month']
df_currency['Category'] = 'USD_KRW Rate'
df_currency['Date'] = pd.to_datetime(df_currency['Date'], format='%M/%d/%y')
df_currency['Date'] = df_currency['Date'].dt.strftime('%Y-%M-%d')
df_currency = df_currency.rename(columns={' Close':'Close'})
df_currency = df_currency.rename(columns={' Close':'Close'})
df_currency = df_currency[['Date','Year','Month','Year_Month','Close','Category']]
df_currency = df_currency.sort_values(['Year','Date'])
df_currency = df_currency.set_index('Date')
df_currency_2019 = df_currency[df_currency['Year'] >= '2019']
df_currency_2008 = df_currency[df_currency['Year'] >= '2008']

df_currency_2008.head()
```
